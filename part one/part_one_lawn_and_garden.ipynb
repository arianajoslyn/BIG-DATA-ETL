# Extract the Amazon Data into Spark DataFrame

```python
# Read in the data from an S3 Bucket
from pyspark import SparkFiles
url = "https://s3.amazonaws.com/amazon-reviews-pds/tsv/amazon_reviews_us_Lawn_and_Garden_v1_00.tsv.gz"
spark.sparkContext.addFile(url)
lawngarden_df = spark.read.csv(SparkFiles.get("amazon_reviews_us_Lawn_and_Garden_v1_00.tsv.gz"), sep="\t", header=True)
lawngarden_df.show()
```

```python
# Get the number of rows in the DataFrame.
lawngarden_df.count()
```

# Transform the Data

## Create the "review_id_table".

```python
from pyspark.sql.functions import to_date
# Create the "review_id_df" DataFrame with the appropriate columns and data types.
review_df = lawngarden_df.select(["review_id", "customer_id", "product_id", "product_parent", to_date("review_date", 'yyyy-MM-dd').alias("review_date")])
review_df.show()
```

## Create the "products" Table

```python
# Create the "products_df" DataFrame that drops the duplicates in the "product_id" and "product_title columns. 
products_df = lawngarden_df.select(["product_id", "product_title"]).drop_duplicates(["product_id"])
products_df.show()
```

## Create the "customers" Table

```python
# Create the "customers_df" DataFrame that groups the data on the "customer_id" by the number of times a customer reviewed a product. 
customers_df = lawngarden_df.groupby("customer_id").agg({"customer_id": "count"}).withColumnRenamed("count(customer_id)", "customer_count")
customers_df.show()
```

## Create the "vine_table".

```python
# Create the "vine_df" DataFrame that has the "review_id", "star_rating", "helpful_votes", "total_votes", and "vine" columns. 
vine_df = lawngarden_df.select(["review_id", "star_rating", "helpful_votes", "total_votes", "vine"])
vine_df.show()
```
